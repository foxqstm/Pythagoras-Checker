```python
import math
import numpy as np

n=10
N=10000
A_B=[[0,0]]
A=3
B=4
cnt2=0

for s in range(1,N+1):
    if(s%100==0):
        print(s)
    for t in range (1,s+1):
        if (((s+t)%2)==0):
            continue
        if (math.gcd(s,t))!=1:
            continue
        A=s**2-t**2
        B=2*s*t

        a_b=[[0,0]]
        cnt=0
        for i in range(1,n+1):
            for j in range (1,n+1):
                Ap=A**i
                Bp=B**j
                ApBp=Ap+Bp
                sqApBp=math.sqrt(ApBp)
                Rootc=math.ceil(sqApBp)
                Rootf=math.floor(sqApBp)
                if (ApBp==Rootc**2) or (ApBp==Rootf**2):
                    a_b.append([i,j])
                cnt+=1

        if(len(a_b)>=4):
            cnt2+=1
            A_B.append([A,B])
            print(A_B[cnt2],a_b,len(a_b)-1)
print(A_B)
```
